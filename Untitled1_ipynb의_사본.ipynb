```python
import random
import numpy as np
import matplotlib.pyplot as plt

Berry = 5400
Black = 0.0996
Crystal = 2980

def simulation1(s, p):
    total = 0
    wonder = 0
    sweet = 0
    d = s + p

    total += getWonderBerry()
    wonder += 1

    while True:
        if wonder == 0:
            total += getWonderBerry()
            wonder += 1
        if sweet == 0:
            if wonder == 0:
                total += getWonderBerry()
                wonder += 1
        else:
            sweet -= 1
        wonder -= 1

        total += Crystal
        temp = random.random()
        if temp < p:
            break
        if temp < d:
            sweet += 1

    return total

def getWonderBerry():
    price = 0
    while random.random() >= Black:
        price += Berry
    return price

def main():
    random.seed()
    results = []
    for i in range(10000):
        result = simulation1(0.864, 0.1164)
        results.append(result)

    results = np.array(results)
    percentiles = np.percentile(results, [10, 25, 50, 75, 90])

    # Plotting the results
    plt.figure(figsize=(10, 6))
    plt.hist(results, bins=50, color='skyblue', edgecolor='gray')
    plt.axvline(x=percentiles[2], color='red', linestyle='--', label='50% Perce')
    plt.xlabel('Percentiles')
    plt.ylabel('Total')
    plt.xticks([10, 25, 50, 75, 90], ['10%', '25%', '50%', '75%', '90%'])
    plt.title('Simulation Results')
    plt.legend()
    plt.show()

    print("Percentiles:")
    for i, percentile in enumerate([10, 25, 50, 75, 90]):
        print(f"{percentile}th percentile: {percentiles[i]}")

if __name__ == '__main__':
    main()

```
